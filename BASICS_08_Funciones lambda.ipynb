<center>  <h1> 08.- FUNCIONES $lambda$ </h1> </center>

Las funciones $lambda$ son funciones de tipo $anónima$, técnicamente carecen de nombre. A diferencia de las funciones instruccionales, en Python, las funciones $lambda$ no se definen con la palabra clave $def$. Este tipo de funciones generalmente se puede utilizar para realizar cálculos numéricos de funciones matemáticas.

Este tipo de funciones normalmente se definen en una línea (instrucción) que incluye la asignación de la función a una variable, la palabra reservada $lambda$, los argumentos de entrada, dos puntos "$:$" y la expresión matemática. 

- $variable\:=\:lambda\:argumentos:\:expresión\:matemática$

Los argumentos de entrada de este tipo de funciones siguen las reglas de las funciones instruccionales $def$, por lo que también pueden operar diferentes tipos de datos y datos múltiples.

“$Python$ $lambdas$ $are$ $only$ $a$ $shorthand$ $notation$ $if$ $you$ $are$ $too$ $lazy$ $to$ $define$ $a$ $function.$”

```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

<h3> ¿Cómo se definen las funciones $lambda$? </h3>

1. Primero se escribe la palabra reservada $lambda$ 
2. Se escriben los $argumentos$ de entrada 
3. Dos puntos
4. Expresión matemática 

A pesar de no tener nombre, este tipo de función se puede asignar a variables y de este modo guardar los resultados en memoria.

```python
# Ejemplo de función lambda: Funciones de seno y coseno

F1 = lambda th: np.sin(th)

F2 = lambda th: np.cos(th)
```

```python
# Creamos una variable de tipo arreglo:
th = np.arange(-5,5,.25)

# Operamos la función lambda y guardamos las salidas en una nueva variable:
salF1 = F1(th)

salF2 = F2(th)
# Imprimimos las salidas:
print(salF1)
print(' ')
print(salF2)
```

De forma auxiliar se puede utilizar la librería $matplotlib.pyplot$ para la representacipon gráfica de las salidas de la función.

```python
# Graficar las salidas como puntos:
plt.plot(th, salF1, ".", markersize=5, color="black") 
plt.plot(th, salF2, ".", markersize=5, color="blue")  

# Graficar las salidas como líneas:
plt.plot(th, salF1, ls="--", linewidth=.5, color="black") 
plt.plot(th, salF2, ls="--", linewidth=.5, color="black")   

plt.xlabel('Valores de theta')   # etiqueta del eje horizontal
plt.ylabel('salF1, salF2')    # etiqueta del eje vertical
plt.show()        # instrucción para mostrar el gráfico
```
